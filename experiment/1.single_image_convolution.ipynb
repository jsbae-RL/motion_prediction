```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
image_path = './dataset/train_data/00479.png'
image = Image.open(image_path).convert('L')
image = np.array(image)
image.shape
```

```python
def custom_convolution(image, kernel):
    """Apply convolution operation manually."""
    img_h, img_w = image.shape
    k_h, k_w = kernel.shape
    
    output_h = img_h - k_h + 1
    output_w = img_w - k_w + 1
    output = np.zeros((output_h, output_w))
    
    for i in range(output_h):
        for j in range(output_w):
            region = image[i:i+k_h, j:j+k_w]
            output[i, j] = np.sum(region * kernel)
    
    return output
```

```python
kernel_v = np.array([[-1, 0, 1], 
                     [-1, 0, 1], 
                     [-1, 0, 1]])

kernel_h = np.array([[-1, -1, -1], 
                     [0, 0, 0], 
                     [1, 1, 1]])
```

```python
# Apply convolution
filtered_image_v = custom_convolution(image, kernel_v)

filtered_image_h = custom_convolution(image, kernel_h)
```

```python
# Display original and filtered images
fig, axes = plt.subplots(1, 4, figsize=(25, 10))
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(filtered_image_v, cmap='gray')
axes[1].set_title("Filtered Image v")
axes[1].axis("off")

axes[2].imshow(filtered_image_h, cmap='gray')
axes[2].set_title("Filtered Image h")
axes[2].axis("off")

axes[3].imshow(fileter, cmap='gray')
axes[3].set_title("Filtered Image h")
axes[3].axis("off")
plt.show()
```
